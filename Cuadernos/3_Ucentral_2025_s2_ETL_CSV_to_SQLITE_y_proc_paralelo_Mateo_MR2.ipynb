```python
#funcion para cargar datos en tablas maestras (departamento, sisben, eps)
def cargar_datos_a_tablas_maestras(df,db_path,columna, tabla_destino):
  conn= sqlite3.connect(db_path)
  cursor=conn.cursor()
  valores_unicos =df[columna].unique()
  for valor in valores_unicos:
    if(pd.notna(valor)): #ignorar valores NaN
      try:
        cursor.execute(f"INSERT OR IGNORE INTO {tabla_destino} (Nombre_{tabla_destino.replace('_','')}) VALUES (?)", (valor,))
      except sqlite3.Error as e:
        print(f"Error al insertar en la tabla {tabla_destino}: {e}")
        pass
  conn.commit()
  conn.close()
```

```python
def revisar_contenido_de_una_tabla(tabla_nombre, whereColumna='',whereValor='', limit=10, order_by_columna=None, order_asc=True):
  conn=sqlite3.connect(db_path)
  cursor=conn.cursor()
  cursor.execute("SELECT COUNT(*) FROM "+tabla_nombre)
  total_registros=cursor.fetchone()[0]
  print(f"Total de registros en la tabla {tabla_nombre}: {total_registros}")
  sql="SELECT * FROM "+tabla_nombre
  params = ()
  if len(whereColumna)>0:
    sql+=" WHERE "+whereColumna+"=?"
    params = (whereValor,)
  if(order_by_columna):
    orderbyTipo="ASC" if order_asc else "DESC"
    sql+=f" ORDER BY {order_by_columna} {orderbyTipo}"
  sql+=" LIMIT "+str(limit)
  print(sql)
  cursor.execute(sql, params)
  resultados=cursor.fetchall()
  for fila in resultados:
    print(fila)
  conn.close()
```

```python
#para cada municipio buscarle su departamento
def cargar_municipios(df,pd_path):
  conn= sqlite3.connect(db_path)
  cursor=conn.cursor()
  tot_creados=0
  for index,row in df.iterrows():
    departamento  =row['Departamento']
    municipio     =row['Municipio']
    cursor.execute("SELECT Id_Departamento FROM Departamento WHERE Nombre_Departamento=?", (departamento,))
    departamento_id=cursor.fetchone()
    if departamento_id:
      departamento_id=departamento_id[0]
      try:
        cursor.execute("INSERT OR IGNORE INTO Municipio (Nombre_Municipio, Id_Departamento) VALUES (?, ?)", (municipio, departamento_id))
        tot_creados+=1
        print(f"creado {tot_creados} => {municipio} en el dpto {departamento} ")
      except sqlite3.sqlite3.IntegrityError:
        pass
    else:
      print(f"ERRROR: Departamento {departamento} para el municipio {municipio} no encontrado")
  conn.commit()
  conn.close()
```

```python
conn= sqlite3.connect(db_path)
cursor=conn.cursor()
cursor.execute('''
CREATE TABLE IF NOT EXISTS Detalle (
    Id_Detalle INTEGER PRIMARY KEY AUTOINCREMENT,
    Id_Municipio INTEGER,
    Id_Sisben INTEGER,
    codigo_Eps TEXT,
    Genero TEXT,
    Grupo_etario TEXT,
    Tipo_afiliado TEXT,
    Estado_afiliado TEXT,
    Condicion_beneficiario TEXT,
    Regimen TEXT,
    Zona_afiliacion TEXT,
    Cantidad_registros INTEGER,
    FOREIGN KEY (Id_Municipio) REFERENCES Municipio(Id_Municipio),
    FOREIGN KEY (Id_Sisben) REFERENCES Sisben(Id_Sisben),
    FOREIGN KEY (codigo_Eps) REFERENCES Eps(codigo_Eps)
)
''')
conn.commit()
conn.close()
```

```python
def cargar_detalle(df,db_path):
  conn= sqlite3.connect(db_path)
  cursor=conn.cursor()
  total_creados=0
  for index,row in df.iterrows():
    municipio   =row['Municipio']
    departamento=row['Departamento']
    #--------para obtener el id del sisben
    if (row['Nivel del Sisbén'] is None):
      id_sisben=0
    else:
      sisben=row['Nivel del Sisbén']
      cursor.execute("SELECT Id_Sisben FROM Sisben WHERE Nombre_Sisben=?", (sisben,))
      sisben_id=cursor.fetchone()
      if sisben_id:
        id_sisben=sisben_id[0]
      else:
        print(f"ERRROR: Sisben {sisben} para el municipio {municipio} no encontrado")
    #----------para obtener id del municipio
    cursor.execute("SELECT Id_Municipio FROM Municipio WHERE Nombre_Municipio=? AND Id_Departamento=(SELECT Id_Departamento FROM Departamento WHERE Nombre_Departamento=?)", (municipio,departamento))
    municipio_id=cursor.fetchone()
    if municipio_id:
      municipio_id=municipio_id[0]
    else:
      print(f"ERRROR: Municipio {municipio} para el departamento {departamento} no encontrado")
    #--------------guarde cada registro en la tabla---
    try:
      cursor.execute('''
      INSERT INTO Detalle (Id_Municipio, Id_Sisben, codigo_Eps, Genero, Grupo_etario, Tipo_afiliado, Estado_afiliado, Condicion_beneficiario, Regimen, Zona_afiliacion, Cantidad_registros)
      VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)''',
                    (municipio_id,
                      id_sisben,
                      row['Código de la entidad'],
                      row['Género'],
                      row['Grupo etario'],
                      row['Tipo de afiliado'],
                      row['Estado del afiliado'],
                      row['Condición del beneficiario'],
                      row['Régimen'],
                      row['Zona de Afiliación'],
                      row['Cantidad de registros']
                    ))
      total_creados+=1
      print(f"creado {total_creados}")
    except sqlite3.IntegrityError as e:
      print(f"No fue posible crear el registro due to {e}")
      pass
  conn.commit()
  conn.close()

```

```python
# =========================================
# 1. Librerías necesarias
# =========================================
!pip install rapidfuzz unidecode openpyxl

from rapidfuzz import process, fuzz
import unidecode
import sqlite3
import pandas as pd


# =========================================
# 2. Diccionario oficial EPS (ADRES 2025)
# =========================================
EPS_OFICIALES = {
    "sanitas": ["sanitas", "eps sanitas", "santias", "santinas salud", "sanitas s.a.s."],
    "nueva eps": ["nueva eps", "nueva", "nueva eps sa", "nueva salud", "nueva e.p.s.", "nueva eps s.a."],
    "sura": ["sura", "eps sura", "sura salud", "eps suramericana", "suramericana de seguros"],
    "compensar": ["compensar", "eps compensar", "compensar salud", "caja de compensación familiar compensar"],
    "famisanar": ["famisanar", "eps famisanar", "eps famisanar s.a.s."],
    "coosalud": ["coosalud", "eps coosalud", "coosalud eps s.a."],
    "mutual ser": ["mutual ser", "eps mutual ser", "asociacion mutual ser empresa solidaria de salud"],
    "salud total": ["salud total", "eps salud total", "salud total entidad promotora de salud"],
    "asmet salud": ["asmet", "asmet salud", "eps asmet", "asmet salud eps s.a.s."],
    "capital salud": ["capital", "capital salud", "eps capital", "capital salud entidad promotora de salud"],
    "emssanar": ["emssanar", "eps emssanar", "emssanar s.a.s."],
    "saviasalud": ["saviasalud", "eps savia salud", "savia salud", "alianza medellin antioquia eps s.a.s. savia salud eps"],
    "aliansalud": ["aliansalud", "aliansalud eps", "aliansalud eps s.a."],
    "sos": ["sos", "eps sos", "sura sos", "servicio occidental de salud s.a. s.o.s.'"],
    "ecoopsos": ["ecoopsos", "eps ecoopsos", "ecoopsos eps s.a.s.'"],
    "dusakawi": ["dusakawi", "eps dusakawi", "asociación de cabildos indígenas del cesar y guajira dusakawi a.r.s.i.'"],
    "indigenas mallamas": ["mallamas", "eps indigenas mallamas", "entidad promotora de salud mallamas epsi'"],
    "pijaos salud": ["pijaos", "pijaos salud", "eps pijaos", "pijaos salud epsi'"],
    "comfachoco": ["comfachoco", "eps comfachoco", "caja de compensación familiar del chocó'"],
    "anase": ["anase", "eps anase", "empresa promotora de salud indígena anas wayuu epsi'"]
}


# =========================================
# 3. Funciones de normalización y mapeo
# =========================================
def normalizar_nombre(nombre: str) -> str:
    """Normaliza un nombre de EPS quitando tildes, eps, salud, y caracteres especiales."""
    if not isinstance(nombre, str):
        return ""
    nombre = nombre.lower()
    nombre = unidecode.unidecode(nombre)
    nombre = nombre.replace("eps", "").replace("salud", "")
    nombre = "".join(c for c in nombre if c.isalnum() or c.isspace())
    return nombre.strip()


def mapear_eps(nombre: str, eps_dict=EPS_OFICIALES, umbral=85) -> str:
    """
    Mapea un nombre de EPS a su nombre oficial usando primero un diccionario
    y luego fuzzy matching.
    """
    nombre_norm = normalizar_nombre(nombre)

    # Buscar coincidencia exacta en el diccionario
    for eps_base, variaciones in eps_dict.items():
        if nombre_norm in [normalizar_nombre(v) for v in variaciones]:
            return eps_base


    coincidencia, score, _ = process.extractOne(
        nombre_norm,
        list(eps_dict.keys()),
        scorer=fuzz.WRatio
    )
    if score >= umbral:
        return coincidencia


    return nombre_norm


# =========================================
# 4. Crear columna EPS_normalizada en df_tutelas_eps and df_eps
# =========================================
df_tutelas_eps["EPS_normalizada"] = df_tutelas_eps["demandado"].apply(mapear_eps)
df_eps['Nombre_Eps_normalizado'] = df_eps['Nombre_Eps'].apply(normalizar_nombre)


df_tutelas_normalized_counts = df_tutelas_eps.groupby("EPS_normalizada").size().reset_index(name='total_tutelas_normalizadas')


df_eps_with_tutelas = pd.merge(df_eps, df_tutelas_normalized_counts, left_on='Nombre_Eps_normalizado', right_on='EPS_normalizada', how='left')


df_eps_with_tutelas = df_eps_with_tutelas.drop(columns='Nombre_Eps_normalizado')
df_eps_with_tutelas['total_tutelas_normalizadas'] = df_eps_with_tutelas['total_tutelas_normalizadas'].fillna(0).astype(int)


display(df_eps_with_tutelas.head())

# =========================================
# 5. Exportar DataFrame a SQLite
# =========================================
conn = sqlite3.connect(db_path)
df_eps_with_tutelas.to_sql("eps_with_tutelas", conn, if_exists="replace", index=False)
conn.close()


# =========================================
# 6. Generar reporte de variaciones (diccionario invertido)
# =========================================
reporte_eps = (
    df_tutelas_eps.groupby("EPS_normalizada")["demandado"]
    .apply(lambda x: sorted(x.unique()))
    .reset_index()
    .rename(columns={"demandado": "Variaciones"})
)

# Guardar en Excel
reporte_eps.to_excel("reporte_diccionario_eps.xlsx", index=False)

print("✅ Proceso completado. Base de datos dbua.db updated with eps_with_tutelas table and reporte_diccionario_eps.xlsx generated")
```

```python
import re
import pandas as pd
from unidecode import unidecode
from rapidfuzz import process, fuzz

# === Función de normalización ===
def normalizar_eps(nombre: str) -> str:
    """
    Normaliza nombres de EPS:
    - pasa a minúsculas
    - elimina tildes
    - quita caracteres especiales
    - compacta espacios
    """
    if pd.isna(nombre):
        return ""
    nombre = unidecode(str(nombre)).lower()
    nombre = re.sub(r'[^a-z0-9\s]', ' ', nombre)   # dejar solo letras/números
    nombre = re.sub(r'\s+', ' ', nombre).strip()
    return nombre

# === Inicializar columna en df_eps ===
df_eps['total_tutelas'] = 0
demandados_encontrados = set()

# Normalizamos demandados en df_tutelas_eps
df_tutelas_eps['demandado_norm'] = df_tutelas_eps['demandado'].apply(normalizar_eps)

# Iterar sobre EPS registradas
for index, row in df_eps.iterrows():
    eps_nombre = row['Nombre_Eps']
    eps_norm = normalizar_eps(eps_nombre)

    # Coincidencia difusa
    candidatos = process.extract(
        eps_norm,
        df_tutelas_eps['demandado_norm'].unique(),
        scorer=fuzz.partial_ratio,
        limit=10
    )

    # Tomar solo los similares con ratio >= 80
    similares = [c[0] for c in candidatos if c[1] >= 80]

    # Buscar tutelas con esos nombres
    matching_tutelas = df_tutelas_eps[df_tutelas_eps['demandado_norm'].isin(similares)]
    total_tutelas = matching_tutelas.shape[0]

    # Guardar en df_eps
    df_eps.loc[index, 'total_tutelas'] = total_tutelas

    # Actualizar lista de demandados encontrados (nombre original)
    demandados_encontrados.update(matching_tutelas['demandado'].tolist())

# EPS demandadas no encontradas
demandados_no_encontrados = set(df_tutelas_eps['demandado'].unique()) - demandados_encontrados

print(f"EPS demandadas encontradas: {len(demandados_encontrados)}")
print(f"EPS demandadas NO encontradas: {len(demandados_no_encontrados)}")

# Mostrar ejemplos de no encontrados
for demandado in list(demandados_no_encontrados)[:30]:
    print(demandado)
```

```python
# =========================================
# 1. Librerías necesarias
# =========================================
!pip install rapidfuzz unidecode openpyxl

from rapidfuzz import process, fuzz
import unidecode
import sqlite3
import pandas as pd
import numpy as np
import re
from collections import defaultdict
import time

# =========================================
# 2. Diccionario oficial EPS completo (Colombia 2024)
# =========================================
EPS_OFICIALES = {
    # EPS del Régimen Contributivo
    "sanitas": {
        "variaciones": ["sanitas", "eps sanitas", "santias", "santinas salud", "sanitas s.a.s.",
                       "sanitas eps", "sanitas s.a", "sanitas colombia", "eps y medicina prepagada sanitas s.a."],
        "codigos": ["800188", "860010"],
        "regimen": "contributivo"
    },
    "nueva eps": {
        "variaciones": ["nueva eps", "nueva", "nueva eps sa", "nueva salud", "nueva e.p.s.",
                       "nueva eps s.a.", "nueva entidad promotora de salud", "nuevaepssa"],
        "codigos": ["800191", "860013"],
        "regimen": "contributivo"
    },
    "sura": {
        "variaciones": ["sura", "eps sura", "sura salud", "eps suramericana", "suramericana de seguros",
                       "sura eps", "seguros sura", "eps sura suramericana", "sura eps suramericana"],
        "codigos": ["800185", "860007"],
        "regimen": "contributivo"
    },
    "compensar": {
        "variaciones": ["compensar", "eps compensar", "compensar salud",
                       "caja de compensación familiar compensar", "compensar eps", "compensar eps salud"],
        "codigos": ["800189", "860011"],
        "regimen": "contributivo"
    },
    "famisanar": {
        "variaciones": ["famisanar", "eps famisanar", "eps famisanar s.a.s.", "famisanar eps",
                       "famisanar s.a.s.", "eps famisanar sas"],
        "codigos": ["800192", "860014"],
        "regimen": "contributivo"
    },
    "coosalud": {
        "variaciones": ["coosalud", "eps coosalud", "coosalud eps s.a.", "coosalud eps",
                       "coosalud s.a.", "corporacion social para la salud coosalud"],
        "codigos": ["800193", "860015"],
        "regimen": "contributivo"
    },
    "mutual ser": {
        "variaciones": ["mutual ser", "eps mutual ser", "asociacion mutual ser empresa solidaria de salud",
                       "mutualser", "mutual ser eps", "asociacion mutual ser"],
        "codigos": ["800194", "860016"],
        "regimen": "contributivo"
    },
    "salud total": {
        "variaciones": ["salud total", "eps salud total", "salud total entidad promotora de salud",
                       "saludtotal", "eps salud total s.a.", "salud total eps"],
        "codigos": ["800195", "860017"],
        "regimen": "contributivo"
    },
    "aliansalud": {
        "variaciones": ["aliansalud", "aliansalud eps", "aliansalud eps s.a.", "alianza salud",
                       "eps aliansalud", "aliansalud s.a."],
        "codigos": ["800196", "860018"],
        "regimen": "contributivo"
    },
    "capital salud": {
        "variaciones": ["capital salud", "capital", "eps capital", "capital salud entidad promotora de salud",
                       "capital salud eps", "eps capital salud"],
        "codigos": ["800197", "860019"],
        "regimen": "contributivo"
    },
    "savia salud": {
        "variaciones": ["saviasalud", "eps savia salud", "savia salud", "alianza medellin antioquia eps s.a.s. savia salud eps",
                       "savia", "eps savia", "savia salud eps", "alianza medellin antioquia"],
        "codigos": ["800198", "860020"],
        "regimen": "contributivo"
    },
    "comfenalco valle": {
        "variaciones": ["comfenalco valle", "eps comfenalco valle", "comfenalco valle eps",
                       "caja de compensacion familiar del valle comfenalco", "comfenalco eps"],
        "codigos": ["800199", "860021"],
        "regimen": "contributivo"
    },

    # EPS del Régimen Subsidiado
    "asmet salud": {
        "variaciones": ["asmet", "asmet salud", "eps asmet", "asmet salud eps s.a.s.",
                       "asmet salud eps", "asmet s.a.s.", "asmet salud sas"],
        "codigos": ["800200", "860022"],
        "regimen": "subsidiado"
    },
    "emssanar": {
        "variaciones": ["emssanar", "eps emssanar", "emssanar s.a.s.", "eps emssanar sas",
                       "emssanar eps", "empresa mutual para la salud de nariño"],
        "codigos": ["800201", "860023"],
        "regimen": "subsidiado"
    },
    "sos": {
        "variaciones": ["sos", "eps sos", "sura sos", "servicio occidental de salud s.a.",
                       "sos eps", "servicio occidental de salud", "sos salud"],
        "codigos": ["800218", "860040"],
        "regimen": "contributivo"
    },
    "ecoopsos": {
        "variaciones": ["ecoopsos", "eps ecoopsos", "ecoopsos eps s.a.s.",
                       "ecoopsos eps", "ecoopsos s.a.s.", "cooperativa ecoopsos"],
        "codigos": ["800219", "860041"],
        "regimen": "contributivo"
    },

    # EPS Indígenas (EPSI)
    "dusakawi": {
        "variaciones": ["dusakawi", "eps dusakawi", "asociación de cabildos indígenas del cesar y guajira dusakawi",
                       "dusakawi epsi", "dusakawi ars i", "asociacion cabildos indigenas cesar guajira"],
        "codigos": ["800207", "860029"],
        "regimen": "indigena"
    },
    "mallamas": {
        "variaciones": ["mallamas", "eps indigenas mallamas", "entidad promotora de salud mallamas epsi",
                       "mallamas epsi", "eps mallamas", "asociacion indigena del putumayo"],
        "codigos": ["800208", "860030"],
        "regimen": "indigena"
    },
    "pijaos salud": {
        "variaciones": ["pijaos", "pijaos salud", "eps pijaos", "pijaos salud epsi",
                       "pijaos epsi", "asociacion indigena del tolima pijaos"],
        "codigos": ["800209", "860031"],
        "regimen": "indigena"
    },
    "anase": {
        "variaciones": ["anase", "eps anase", "empresa promotora de salud indígena anas wayuu epsi",
                       "anase wayuu", "anas wayuu", "anase epsi"],
        "codigos": ["800210", "860032"],
        "regimen": "indigena"
    },

    # Otras EPS importantes
    "comfachoco": {
        "variaciones": ["comfachoco", "eps comfachoco", "caja de compensación familiar del chocó",
                       "comfachoco eps", "comfachoco caja compensacion"],
        "codigos": ["800220", "860042"],
        "regimen": "contributivo"
    },
    "coomeva": {
        "variaciones": ["coomeva", "eps coomeva", "coomeva eps",
                       "cooperativa medica coomeva", "coomeva salud"],
        "codigos": ["800217", "860039"],
        "regimen": "contributivo"
    },
    "medimas": {
        "variaciones": ["medimas", "eps medimas", "medimas eps",
                       "medimas salud", "eps medimas salud"],
        "codigos": ["800227", "860049"],
        "regimen": "subsidiado"
    }
}

# Diccionario de códigos NIT para matching adicional
EPS_CODIGOS_NIT = {
    "800188": "sanitas", "860010": "sanitas",
    "800191": "nueva eps", "860013": "nueva eps",
    "800185": "sura", "860007": "sura",
    "800189": "compensar", "860011": "compensar",
    "800192": "famisanar", "860014": "famisanar",
    "800193": "coosalud", "860015": "coosalud",
    "800194": "mutual ser", "860016": "mutual ser",
    "800195": "salud total", "860017": "salud total",
    "800196": "aliansalud", "860018": "aliansalud",
    "800197": "capital salud", "860019": "capital salud",
    "800198": "savia salud", "860020": "savia salud",
    "800199": "comfenalco valle", "860021": "comfenalco valle",
    "800200": "asmet salud", "860022": "asmet salud",
    "800201": "emssanar", "860023": "emssanar",
    "800218": "sos", "860040": "sos",
    "800219": "ecoopsos", "860041": "ecoopsos",
    "800207": "dusakawi", "860029": "dusakawi",
    "800208": "mallamas", "860030": "mallamas",
    "800209": "pijaos salud", "860031": "pijaos salud",
    "800210": "anase", "860032": "anase",
    "800220": "comfachoco", "860042": "comfachoco",
    "800217": "coomeva", "860039": "coomeva",
    "800227": "medimas", "860049": "medimas"
}

# =========================================
# 3. Funciones de normalización mejoradas
# =========================================
def normalizar_nombre_avanzado(nombre: str) -> str:
    """Normalización más robusta de nombres de EPS"""
    if not isinstance(nombre, str) or pd.isna(nombre):
        return ""

    # Convertir a minúsculas y quitar tildes
    nombre = nombre.lower().strip()
    nombre = unidecode.unidecode(nombre)

    # Remover patrones comunes no informativos
    patrones_remover = [
        r'\beps\b', r'\bsalud\b', r'\bs\.a\.\b', r'\bs\.a\.s\.\b',
        r'\be\.p\.s\.\b', r'\bentidad\b', r'\bpromotora\b',
        r'\bde\b', r'\bla\b', r'\bel\b', r'\blos\b', r'\blas\b'
    ]

    for patron in patrones_remover:
        nombre = re.sub(patron, '', nombre)

    # Remover caracteres especiales y espacios múltiples
    nombre = re.sub(r'[^a-z0-9\s]', '', nombre)
    nombre = re.sub(r'\s+', ' ', nombre).strip()

    return nombre

def extraer_patrones_comunes(texto):
    """Extrae patrones comunes para mejorar el matching"""
    if not isinstance(texto, str):
        return {'palabras_clave': []}

    patrones = {
        'palabras_clave': re.findall(r'\b(sanitas|sura|nueva|compensar|famisanar|coosalud|mutual|salud total|asmet|emssanar|savia|aliansalud|capital|sos|ecoopsos|dusakawi|mallamas|pijaos|anase|comfachoco|coomeva|medimas)\b', texto.lower())
    }
    return patrones

def mapear_por_codigo(codigo: str) -> str:
    """Mapea códigos NIT a nombres de EPS"""
    if pd.isna(codigo) or not str(codigo).strip():
        return None

    codigo_str = str(codigo).strip()
    return EPS_CODIGOS_NIT.get(codigo_str, None)

# =========================================
# 4. Sistema de matching mejorado
# =========================================
class EPSMatcher:
    def __init__(self, eps_dict, codigos_dict):
        self.eps_dict = eps_dict
        self.codigos_dict = codigos_dict

        # Precomputar todas las variaciones normalizadas para fuzzy matching
        self.variaciones_norm = {}
        self.todas_variaciones = []

        for eps, data in eps_dict.items():
            for variacion in data["variaciones"]:
                variacion_norm = normalizar_nombre_avanzado(variacion)
                self.variaciones_norm[variacion_norm] = eps
                self.todas_variaciones.append(variacion_norm)

        # Añadir también los nombres base de las EPS
        for eps in eps_dict.keys():
            eps_norm = normalizar_nombre_avanzado(eps)
            self.variaciones_norm[eps_norm] = eps
            self.todas_variaciones.append(eps_norm)

        # Eliminar duplicados
        self.todas_variaciones = list(set(self.todas_variaciones))

    def mapear_eps_optimizado(self, nombre: str, umbral_exacto=95, umbral_fuzzy=85) -> str:
        """
        Sistema de matching en cascada para mejor precisión y performance
        """
        if not nombre or pd.isna(nombre):
            return "no_identificada"

        nombre_norm = normalizar_nombre_avanzado(nombre)

        if not nombre_norm:
            return "no_identificada"

        # 1. Búsqueda exacta en variaciones normalizadas
        if nombre_norm in self.variaciones_norm:
            return self.variaciones_norm[nombre_norm]

        # 2. Búsqueda por patrones específicos
        patrones = extraer_patrones_comunes(nombre)
        for palabra in patrones['palabras_clave']:
            for eps in self.eps_dict.keys():
                if palabra in eps:
                    return eps

        # 3. Fuzzy matching optimizado
        try:
            coincidencia, score, _ = process.extractOne(
                nombre_norm,
                self.todas_variaciones,
                scorer=fuzz.WRatio,
                processor=None
            )

            if score >= umbral_fuzzy:
                return self.variaciones_norm[coincidencia]
        except:
            pass

        return f"pendiente_clasificacion: {nombre_norm}"

    def mapear_eps_completo(self, nombre: str, codigo: str = None, umbral_exacto=95, umbral_fuzzy=85) -> str:
        """
        Sistema de matching completo que usa nombre y código
        """
        # Primero intentar por código si está disponible
        if codigo:
            eps_por_codigo = mapear_por_codigo(codigo)
            if eps_por_codigo:
                return eps_por_codigo

        # Si no hay código o no se encontró, usar el nombre
        return self.mapear_eps_optimizado(nombre, umbral_exacto, umbral_fuzzy)

# =========================================
# 5. Procesamiento por lotes para grandes volúmenes
# =========================================
def procesar_por_lotes(df, columna_entrada, funcion_mapeo, tamaño_lote=1000, **kwargs):
    """Procesa grandes datasets por lotes para mejor performance"""
    resultados = []
    total_registros = len(df)

    for i in range(0, total_registros, tamaño_lote):
        lote = df.iloc[i:i+tamaño_lote]
        if 'codigo' in kwargs and kwargs['codigo'] in df.columns:
            resultados_lote = lote.apply(
                lambda x: funcion_mapeo(x[columna_entrada], x[kwargs['codigo']]),
                axis=1
            )
        else:
            resultados_lote = lote[columna_entrada].apply(funcion_mapeo)

        resultados.extend(resultados_lote)

        # Progress tracking
        if (i // tamaño_lote) % 10 == 0:
            print(f"Procesados {min(i+tamaño_lote, total_registros)}/{total_registros} registros")

    return resultados

# =========================================
# 6. Análisis de calidad y reporting
# =========================================
def generar_reporte_calidad(df, columna_original, columna_mapeada):
    """Genera reporte detallado de la calidad del mapeo"""
    total = len(df)
    mapeados = sum(~df[columna_mapeada].str.startswith('pendiente_clasificacion'))
    no_identificados = sum(df[columna_mapeada] == "no_identificada")

    reporte = {
        'total_registros': total,
        'mapeados_correctamente': mapeados,
        'no_identificados': no_identificados,
        'pendiente_clasificacion': total - mapeados - no_identificados,
        'porcentaje_exito': (mapeados / total) * 100 if total > 0 else 0,
        'distribucion_eps': df[columna_mapeada].value_counts().head(20).to_dict(),
        'ejemplos_no_mapeados': df[df[columna_mapeada].str.startswith('pendiente_clasificacion')][columna_original].head(20).tolist() if any(df[columna_mapeada].str.startswith('pendiente_clasificacion')) else []
    }

    return reporte

# =========================================
# 7. Implementación principal
# =========================================
def main():
    # Inicializar el matcher
    matcher = EPSMatcher(EPS_OFICIALES, EPS_CODIGOS_NIT)

    # Procesar datos en lotes
    print("Iniciando procesamiento de datos...")
    start_time = time.time()

    # Verificar si existe columna de código para matching mejorado
    tiene_codigo = 'codigo_eps' in df_tutelas_eps.columns

    if tiene_codigo:
        print("✅ Usando matching combinado (nombre + código)")
        df_tutelas_eps["EPS_normalizada"] = procesar_por_lotes(
            df_tutelas_eps, "demandado", matcher.mapear_eps_completo, codigo='codigo_eps'
        )
    else:
        print("⚠️  Usando solo matching por nombre (no se encontró columna 'codigo_eps')")
        df_tutelas_eps["EPS_normalizada"] = procesar_por_lotes(
            df_tutelas_eps, "demandado", matcher.mapear_eps_optimizado
        )

    # Procesar df_eps
    df_eps['Nombre_Eps_normalizado'] = df_eps['Nombre_Eps'].apply(normalizar_nombre_avanzado)

    # Generar reporte de calidad
    reporte = generar_reporte_calidad(df_tutelas_eps, "demandado", "EPS_normalizada")

    print(f"✅ Procesamiento completado en {time.time() - start_time:.2f} segundos")
    print(f"📊 Estadísticas:")
    print(f"   - Total registros: {reporte['total_registros']}")
    print(f"   - Mapeados correctamente: {reporte['mapeados_correctamente']} ({reporte['porcentaje_exito']:.2f}%)")
    print(f"   - No identificados: {reporte['no_identificados']}")
    print(f"   - Pendientes de clasificación: {reporte['pendiente_clasificacion']}")

    # Continuar con el proceso de merging y exportación...
    df_tutelas_normalized_counts = df_tutelas_eps.groupby("EPS_normalizada").size().reset_index(name='total_tutelas_normalizadas')

    df_eps_with_tutelas = pd.merge(
        df_eps,
        df_tutelas_normalized_counts,
        left_on='Nombre_Eps_normalizado',
        right_on='EPS_normalizada',
        how='left'
    )

    df_eps_with_tutelas = df_eps_with_tutelas.drop(columns='Nombre_Eps_normalizado')
    df_eps_with_tutelas['total_tutelas_normalizadas'] = df_eps_with_tutelas['total_tutelas_normalizadas'].fillna(0).astype(int)

    display(df_eps_with_tutelas.head())

    # =========================================
    # 8. Exportar DataFrame a SQLite
    # =========================================
    conn = sqlite3.connect(db_path)
    df_eps_with_tutelas.to_sql("eps_with_tutelas", conn, if_exists="replace", index=False)
    conn.close()

    # =========================================
    # 9. Generar reportes detallados
    # =========================================
    # Reporte de variaciones
    reporte_eps = (
        df_tutelas_eps.groupby("EPS_normalizada")["demandado"]
        .apply(lambda x: sorted(x.unique()[:10]))  # Limitar a 10 ejemplos por EPS
        .reset_index()
        .rename(columns={"demandado": "Variaciones"})
    )

    reporte_eps.to_excel("reporte_diccionario_eps.xlsx", index=False)

    # Guardar registros no clasificados para análisis posterior
    no_clasificados = df_tutelas_eps[df_tutelas_eps["EPS_normalizada"].str.startswith('pendiente_clasificacion')]
    if len(no_clasificados) > 0:
        no_clasificados.to_excel("registros_no_clasificados.xlsx", index=False)
        print(f"📝 {len(no_clasificados)} registros no clasificados guardados en 'registros_no_clasificados.xlsx'")

    # Reporte de calidad
    with pd.ExcelWriter("reporte_calidad_clasificacion.xlsx") as writer:
        pd.DataFrame([reporte]).to_excel(writer, sheet_name='Resumen', index=False)
        df_tutelas_eps["EPS_normalizada"].value_counts().reset_index().to_excel(
            writer, sheet_name='Distribucion_EPS', index=False
        )

    print("✅ Proceso completado. Archivos generados:")
    print("   - reporte_diccionario_eps.xlsx")
    print("   - reporte_calidad_clasificacion.xlsx")
    if len(no_clasificados) > 0:
        print("   - registros_no_clasificados.xlsx")

# Ejecutar el proceso principal
if __name__ == "__main__":
    main()
```
