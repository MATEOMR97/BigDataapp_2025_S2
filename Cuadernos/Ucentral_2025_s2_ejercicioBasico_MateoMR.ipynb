#Categorización de Placas para Vehículos en Colombia


```python
# Categorización variable PLACA para identificar si pertenece a vehiculo o ciclomotor
def categorize_plate(plate):
    if pd.isna(plate):
        return 'Desconocido'
    plate_str = str(plate).strip()
    if len(plate_str) == 6:
        return 'Vehículo'  # Carros (3 letras + 3 números)
    elif len(plate_str) == 5:
        return 'Moto' # Motocicletas (3 letras + 2 números)
    else:
        return 'Otros' # Other formats

df_Santander['TIPO_VEHICULO'] = df_Santander['PLACA'].apply(categorize_plate)

# Display the first few rows with the new column
print(df_Santander[['PLACA', 'TIPO_VEHICULO']].head(20).to_string(index=False))

# Check the distribution of the new category
print("\nDistribución por tipo de vehículo:")
print(df_Santander['TIPO_VEHICULO'].value_counts())
```

```python
import pandas as pd
import re

def categorizar_placa_colombiana(placa):
    """
    Categoriza placas colombianas según los formatos oficiales
    """
    if pd.isna(placa):
        return 'Desconocido'

    placa_str = str(placa).strip().upper()

    # Limpiar la placa: remover espacios, guiones y caracteres especiales
    placa_limpia = re.sub(r'[^A-Z0-9]', '', placa_str)

    # Verificar diferentes formatos de placas colombianas
    if len(placa_limpia) == 6:
        # Formato tradicional: AAA123 (3 letras + 3 números) - Vehículos
        if re.match(r'^[A-Z]{3}\d{3}$', placa_limpia):
            return 'Vehículo'
        # Nuevo formato: ABC12D (3 letras + 2 números + 1 letra) - Vehículos
        elif re.match(r'^[A-Z]{3}\d{2}[A-Z]$', placa_limpia):
            return 'Vehículo'

    elif len(placa_limpia) == 5:
        # Formato tradicional: ABC12 (3 letras + 2 números) - Motos antiguas
        if re.match(r'^[A-Z]{3}\d{2}$', placa_limpia):
            return 'Moto'
        # Nuevo formato: A12BC (1 letra + 2 números + 2 letras) - Motos nuevas
        elif re.match(r'^[A-Z]\d{2}[A-Z]{2}$', placa_limpia):
            return 'Moto'

    elif len(placa_limpia) == 7:
        # Vehículos oficiales o especiales
        if re.match(r'^[A-Z]{3}\d{4}$', placa_limpia) or re.match(r'^[A-Z]{2}\d{3}[A-Z]{2}$', placa_limpia):
            return 'Vehículo Especial'

    # Si no coincide con ningún formato conocido
    return 'Otros'

# Aplicar la función a tu DataFrame
df_Santander['TIPO_VEHICULO'] = df_Santander['PLACA'].apply(categorizar_placa_colombiana)

# Mostrar ejemplos de las primeras filas
print("Primeras 20 filas con la categorización:")
print(df_Santander[['PLACA', 'TIPO_VEHICULO']].head(20).to_string(index=False))

# Análisis de la distribución
print("\n" + "="*60)
print("DISTRIBUCIÓN POR TIPO DE VEHÍCULO")
print("="*60)
distribucion = df_Santander['TIPO_VEHICULO'].value_counts()
print(distribucion)

print("\n" + "="*60)
print("ESTADÍSTICAS DETALLADAS")
print("="*60)
total_registros = len(df_Santander)
desconocidos = distribucion.get('Desconocido', 0)

print(f"Total de registros: {total_registros}")
print(f"Registros con placa desconocida: {desconocidos}")
print(f"Porcentaje de datos válidos: {(total_registros - desconocidos) / total_registros * 100:.2f}%")

# Mostrar ejemplos de formatos inusuales
otros_formatos = df_Santander[df_Santander['TIPO_VEHICULO'] == 'Otros']['PLACA'].dropna()
if not otros_formatos.empty:
    print(f"\nFormatos inusuales encontrados ({len(otros_formatos)}):")
    for placa in otros_formatos.unique()[:10]:  # Mostrar solo los primeros 10 únicos
        print(f"  - '{placa}'")
```

# Gráfica de Distribucuón

# Análisis (%)  PAGADAS Y NO PAGADAS VS TIPO DE VEHÍCULO

```python
import matplotlib.pyplot as plt
import pandas as pd

# Filtrar solo Vehículo y Moto
filtered_df = df_Santander[df_Santander['TIPO_VEHICULO'].isin(['Vehículo', 'Moto'])].copy()

# Crear tabla cruzada con porcentajes
cross_tab = pd.crosstab(filtered_df['TIPO_VEHICULO'],
                        filtered_df['PAGADO_SI_NO'],
                        normalize='index') * 100

# Crear gráfica de barras apiladas
plt.figure(figsize=(10, 6))
ax = cross_tab.plot(kind='bar', stacked=True, color=['#ff6b6b', '#4ecdc4'], edgecolor='black')

plt.title('Porcentaje de multas pagadas y no pagadas por tipo de vehículo\nSantander',
          fontsize=14, fontweight='bold', pad=20)
plt.xlabel('Tipo de Vehículo', fontsize=12, fontweight='bold')
plt.ylabel('Porcentaje (%)', fontsize=12, fontweight='bold')

# Personalizar leyenda
plt.legend(['No Pagadas', 'Pagadas'], title='Estado de Pago',
           bbox_to_anchor=(1.05, 1), loc='upper left')

# Añadir etiquetas de porcentaje en cada segmento
for container in ax.containers:
    # Obtener las etiquetas y posicionarlas en el centro de cada segmento
    labels = [f'{value:.1f}%' if value > 5 else '' for value in container.datavalues]
    ax.bar_label(container, labels=labels, label_type='center', fontsize=10, fontweight='bold')

# Ajustar ejes
plt.xticks(rotation=0)
plt.ylim(0, 100)

# Mostrar gráfica
plt.tight_layout()
plt.show()

# Mostrar tabla con los porcentajes
print("PORCENTAJES DE PAGO POR TIPO DE VEHÍCULO:")
print("=" * 50)
print(cross_tab.round(1))
```
